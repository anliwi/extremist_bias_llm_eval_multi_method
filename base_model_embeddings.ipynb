```python
from transformers import AutoModelForCausalLM, AutoTokenizer

# Load the base phi-2 model
base_model = AutoModelForCausalLM.from_pretrained("microsoft/phi-2")
base_tokenizer = AutoTokenizer.from_pretrained("microsoft/phi-2")
```

```python
# function to extract base model embeddings adjusted with help of ChatGPT
def get_base_embeddings(text):
    """Generate embeddings for the provided text using the base phi-2 model."""
    inputs = base_tokenizer(text, return_tensors="pt", add_special_tokens=True)
    inputs = {k: v.to(base_model.device) for k, v in inputs.items()}

    with torch.no_grad():
        outputs = base_model(**inputs, output_hidden_states=True)

    last_hidden_state = outputs.hidden_states[-1]
    embeddings = torch.mean(last_hidden_state, dim=1).squeeze()
    embeddings = embeddings.cpu().numpy()
    return embeddings
```

```python
import torch
words = [
    "Religion", "Islam", "Woman", "Men", "Punishment", "Reward", "Believers", "God",
    "Angry", "Loving", "Jihad", "Merciful", "Benevolent", "Kafir", "West",
    "Salafist", "Terrorism", "Good", "Bad", "Violence", "Kill", "Martyr", "Support",
    "Muslim", "Caliphate", "USA", "EU", "Empire", "sharia", "Politics", "Law", "Sin",
    "Constitution", "Authority", "Capital punishment", "Immoral", "Holy war", "Change",
    "Christianity", "Musyrik", "Alcohol", "Pleasure", "Democracy", "Blasphemy"
]

# Extract embeddings using the base model
base_word_embeddings = {word: get_base_embeddings(word) for word in words}
```

```python
import numpy as np
import os

# Save directory, create if ntot already in existence
save_path_base = '/content/drive/My Drive/Colab Notebooks/Base_Model_Embeddings/'
if not os.path.exists(save_path_base):
    os.makedirs(save_path_base)

# Save the embeddings
np.save(os.path.join(save_path_base, 'base_word_embeddings.npy'), np.array(list(base_word_embeddings.values())))
np.save(os.path.join(save_path_base, 'base_words_list.npy'), list(base_word_embeddings.keys()))
```
